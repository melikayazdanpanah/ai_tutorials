## 5. Reading Time Series with Pandas
We can read time series data into a Pandas DataFrame.  
Each datapoint has a timestamp (often dates). Multiple values can share the same timestamp.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data (replace with your own CSVs)
data = pd.DataFrame({
    "timestamp": pd.date_range("2020-01-01", periods=1500, freq="D"),
    "data_values": range(1500)
})
data2 = pd.DataFrame({
    "timestamp": pd.date_range("2020-01-01", periods=1500, freq="H"),
    "data_values": range(1500)
})

# Plot the time series in each dataset
fig, axs = plt.subplots(2, 1, figsize=(5, 10))
data.iloc[:1000].plot(y='data_values', ax=axs[0], title="Dataset 1")
data2.iloc[:1000].plot(y='data_values', ax=axs[1], title="Dataset 2")
plt.show()
```

## 7. Getting to Know Our Data
We’ll use two datasets from Kaggle:
1. **Heartbeat Acoustic Data** (audio signals)  
2. **New York Stock Exchange (NYSE) data** (financial time series)  


```python
import librosa as lr
from glob import glob
import numpy as np

# Example path (replace with your folder path containing wav files)
data_dir = "your_audio_folder"

# List all the wav files in the folder
audio_files = glob(data_dir + '/*.wav')

# Read in the first audio file, create the time array
audio, sfreq = lr.load(audio_files[0])
time = np.arange(0, len(audio)) / sfreq

# Plot audio over time
fig, ax = plt.subplots()
ax.plot(time, audio)
ax.set(xlabel='Time (s)', ylabel='Sound Amplitude', title="Heartbeat Audio Example")
plt.show()
```
